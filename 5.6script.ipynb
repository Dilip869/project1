```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("/content/drive/MyDrive/Project1_processed/Chapter5.xlsx","5.6")
df.head(10)
```

```python
df.dropna(thresh=abs(df.shape[1]*0.5), axis=0, inplace=True)
df.dropna(how="all", axis=1, inplace=True)
df.head(5)
```

```python
df.drop(index=[df.index[0],df.index[2]], inplace=True)
df.head(5)
```

```python
df.drop(columns=[df.columns[0],df.columns[3],df.columns[6]], inplace=True)
df.head(5)
```

```python
df.columns=df.iloc[0,:].values
df.head(5)
```

```python
df.drop(index=df.index[0], inplace=True)
df.head(5)
```

```python
df.rename(columns={df.columns[0]:"Total Women MPs",
                   df.columns[2]:"Women First Time Elected",
                   df.columns[3]:"Total First Time Elected",
                   df.columns[4]:"state"}, inplace=True)
df.head(5)
```

```python
melt_df=df.melt(id_vars=df.columns[4],var_name="category",value_name="value")
melt_df.head(5)
```

```python

melt_df=melt_df.assign(units="value in absolte number")
melt_df=melt_df.assign(data_as_on="03-01-2020")
melt_df.head(10)
```

```python
melt_df=melt_df[["data_as_on","state","category","value","units"]]
```

```python
new_df=pd.read_excel('/content/drive/MyDrive/Project1_processed/data_dictionaries.xlsx')
new_df.head(5)
```

```python
ref_df=pd.DataFrame(melt_df["state"].unique())
ref_df.head(5)
```

```python
list1=new_df["state"].to_list()
list2=ref_df[ref_df.columns[0]].to_list()
```

```python
from thefuzz import process
from thefuzz import fuzz
```

```python
result={}
for i in list2:
  result[i]=process.extractOne(i,list1,scorer=fuzz.ratio)[0]
```

```python
melt_df["state"]=melt_df["state"].map(result)
```

```python
melt_df.head(10)
```

```python
melt_df.sort_values(by=["state","category"], inplace=True)
melt_df.head(10)
```

```python
melt_df["value"]=melt_df["value"].apply(lambda x: np.nan if "-" in str(x) else x)
melt_df.head(10)
```

```python
melt_df = melt_df.applymap(lambda x : x.strip() if isinstance(x,str) else x)
melt_df.head(5)
```
