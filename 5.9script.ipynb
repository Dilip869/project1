```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('/content/drive/MyDrive/Project1_processed/Chapter5.xlsx','5.9')
df.head(10)
```

```python
df.dropna(thresh=abs(df.shape[1]*0.5), axis=0, inplace=True)
df.dropna(how="all", axis=1, inplace=True)
df.head(5)
```

```python
df.drop(columns=df.columns[0], inplace=True)
df.head(5)
```

```python
df.columns=df.iloc[0,:].values
df.head(5)
```

```python
df.drop(index=[df.index[0],df.index[1]], inplace=True)
df.head(5)
```

```python
melt_df=df.melt(id_vars=df.columns[3], var_name="fiscal_year", value_name="value")
melt_df.head(5)
```

```python
melt_df.rename(columns={melt_df.columns[0]:"state",melt_df.columns[2]:"value"}, inplace=True)
melt_df.head(5)
```

```python
melt_df=melt_df.assign(units="value in absolute number per 1000 person")
```

```python
melt_df=melt_df[["fiscal_year","state","value","units"]]
melt_df.head(10)
```

```python
new_df=pd.read_excel('/content/drive/MyDrive/Project1_processed/data_dictionaries.xlsx')
new_df.head(5)
```
