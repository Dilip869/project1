```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('/content/drive/MyDrive/Project1_processed/Chapter5.xlsx','5.8')
df.head(10)
```

```python
df.dropna(thresh=abs(df.shape[1]*0.5),axis=0, inplace=True)
df.dropna(how="all", axis=1, inplace=True)
df.head(5)
```

```python
df.drop(columns=[df.columns[0],df.columns[6]], inplace=True)
df.head(5)
```

```python
df.drop(index=[df.index[0],df.index[2]], inplace=True)
```

```python
df.columns=df.iloc[0,:].values
```

```python
df.drop(index=df.index[0], inplace=True)
df.head(5)
```

```python
melt_df=df.melt(id_vars=df.columns[5], var_name="category_of_judges",value_name="value")
melt_df.head(5)
```

```python
melt_df.rename(columns={melt_df.columns[0]:"court"}, inplace=True)
melt_df.head(10)
```

```python
melt_df=melt_df.assign(date_as_on="25-11-2021")
melt_df=melt_df.assign(units="value in absolute numbers")
```

```python
melt_df["note"]=melt_df["category_of_judges"].apply(lambda x: "category_of_judges:Female/Male figures based on total no. of judges" if "Male" in x else ("category_of_judges:Female/Male figures based on total no. of judges" if "Female" in x else"" ))


```

```python
melt_df=melt_df[["date_as_on","court","category_of_judges","value","units","note"]]
melt_df.head(5)
```

```python
melt_df.sort_values(by=["court"], inplace=True)
melt_df.head(10)
```
