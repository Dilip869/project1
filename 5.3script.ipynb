```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('/content/drive/MyDrive/Project1_processed/Chapter5.xlsx','5.3')
df.head(10)
```

```python
df.dropna(thresh=abs(df.shape[1]*0.5), axis=0, inplace=True)
df.dropna(how="all", axis=1, inplace=True)
df.head(10)
```

```python
df.drop(index=[df.index[0],df.index[2]],inplace=True)
df.drop(columns=df.columns[0], inplace=True)
df.head(10)
```

```python
df.columns=df.iloc[0,:].values
```

```python
df.drop(index=df.index[0],inplace=True)
df.head(10)
```

```python
df.drop(columns=[df.columns[2],df.columns[5]], inplace=True)
df.head(10)
```

```python
melt_df=df.melt(id_vars=df.columns[4],var_name="electors_voters_type",value_name="value")
melt_df.head(10)
```

```python
melt_df=melt_df.assign(units="value in thousands")
melt_df.head(10)
```

```python

melt_df.rename(columns={melt_df.columns[0]: "state"}, inplace=True)
melt_df.head(5)
```

```python
melt_df=melt_df.assign(year=2019)
```

```python
melt_df=melt_df[["year","state","electors_voters_type","value","units"]]
melt_df.head(5)
```
