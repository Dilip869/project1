```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('/content/drive/MyDrive/Project1_processed/Chapter5.xlsx','5.12')
df.head(5)
```

```python
df.dropna(thresh=abs(df.shape[1]*0.5), axis=0, inplace=True)
df.dropna(how="all", axis=1, inplace=True)
df.head(10)
```

```python
df.drop(columns=[df.columns[0],df.columns[3]], inplace=True)
df.head(20)
```

```python
df.drop(index=[df.index[0],df.index[2],df.index[7],df.index[17],df.index[18]], inplace=True)
df.head(20)
```

```python
df.columns=df.iloc[0,:].values
df.head(5)
```

```python
df.drop(index=df.index[0], inplace=True)
df.head(5)
```

```python
df.rename(columns={df.columns[0]: "total_police_strength", df.columns[1]:"female_police_strength",df.columns[2]:"police_organization"}, inplace=True)
df.head(15)
```

```python
melt_df=df.melt(id_vars=df.columns[2], var_name="police_strength", value_name="value")
melt_df.head(5)
```

```python
melt_df=melt_df.assign(date_as_on="01-01-2020")
melt_df=melt_df.assign(units="value in absolute number")
melt_df.tail(10)
```

```python
note_dict={"*":"NDRF is a 100% Deputationist Force","#":"NSG is a 100% Deputationist Force, No Reservation percentage has been fixed for women in NSG."}
```

```python
for each_note in note_dict:
  melt_df[f"{each_note}"]=melt_df["police_organization"].apply(lambda x:f"police_organization:{note_dict[each_note]}" if f"{each_note}" in x else "")
melt_df.tail(10)
```

```python
for each_note in note_dict:
  melt_df["police_organization"] = melt_df["police_organization"].apply(lambda x :x.strip(f'{each_note}' ))
melt_df.tail(10)
```

```python
melt_df["note"]=melt_df["*"]+melt_df["#"]
melt_df.head(5)
```

```python
melt_df.drop(columns=['#','*'], axis=1, inplace=True)
melt_df.head(10)
```

```python
melt_df=melt_df[["date_as_on", "police_organization","police_strength","value","units","note"]]
melt_df.head(10)
```

```python
melt_df.sort_values(by=["police_organization"], inplace=True)
melt_df.tail(15)
```

```python
melt_df = melt_df.applymap(lambda x : x.strip() if isinstance(x,str) else x)
```

```python
melt_df.to_csv('/content/drive/MyDrive/Project1_processed/Table 5.12/output.csv', index=False)
```
