```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('/content/drive/MyDrive/Project1_processed/Chapter5.xlsx','5.2')
df.head(10)
```

```python
df.drop(columns=df.columns[0], inplace=True)
df.head(10)
```

```python
df.set_index(df.columns[7], inplace=True)
df.head(10)
```

```python
df = df.applymap(lambda x : x.strip() if isinstance(x,str) else x)
df.head(5)

```

```python
df.iloc[2]=df.iloc[2,:].ffill()
df.head(10)
```

```python
new_columns=pd.MultiIndex.from_arrays([df.iloc[2].values,df.iloc[4].values])
new_columns
```

```python
df.drop(index=df.index[:6], inplace=True)
df.head(5)
```

```python
df.columns=new_columns
df.head(10)
```

```python
df.reset_index(drop=False, inplace=True)
df.head(5)
```

```python
df.set_index([df.columns[0],df.columns[1]],drop=True, inplace=True)
df.head(5)
```

```python
stack_df=df.stack(level=1).reset_index()
stack_df.head(10)
```

```python
stack_df.rename(columns={stack_df.columns[0]:"lok_sabha_number",stack_df.columns[1]:"year",stack_df.columns[2]:"gender","Percentage of electors participating in the election":"electors_participated","Total number of  electors\n(Figures in Millions)":"total_electors"}, inplace=True)
stack_df.head(5)
```

```python

stack_df=stack_df.assign(units="total_electors in Millions, electors_participated in percentage")
stack_df.head(5)
```

```python
statements=[("**", "total includes others."),("*","Calculated on the basis of vaild votes polled.")]
```

```python
for note,comment in statements:
  stack_df["note_1"]=stack_df["electors_participated"].apply(lambda x:f"electors_participated:{comment}" if f"{note}" in str(x) else "")
  stack_df["note_2"]=stack_df["total_electors"].apply(lambda x:f"total_electors:{comment}" if f"{note}" in str(x) else "")
stack_df.head(10)
```

```python
for each_note in statements:
  stack_df["electors_participated"] = stack_df["electors_participated"].apply(lambda x :str(x).strip(f'{each_note}'))
  stack_df["total_electors"] = stack_df["total_electors"].apply(lambda x :str(x).strip(f'{each_note}'))
stack_df.head(10)
```

```python
text="Sex-wise break up of electors is not available for the first, second and fifth General Elections."
note_dict={"First":text,"Second":text,"Fifth":text}
```

```python
for each_note in note_dict:
  stack_df[f"{each_note}"]=stack_df["lok_sabha_number"].apply(lambda x: f"lok_sabha_number:{note_dict[each_note]}" if f"{each_note}" in x else "")
stack_df.head(10)
```

```python
stack_df["note"]=stack_df["note_1"]+stack_df["note_2"]+stack_df["First"]+stack_df["Second"]+stack_df["Fifth"]
stack_df.head(10)
```

```python
stack_df.drop(["note_1","note_2","First","Second","Fifth"], axis=1, inplace=True)
stack_df.head(10)
```

```python
stack_df["electors_participated"]=stack_df["electors_participated"].apply(lambda x:"" if "-" in x else x)
stack_df.head(10)
```

```python
stack_df["total_electors"]=stack_df["total_electors"].apply(lambda x:"" if "-" in x else x)
stack_df.head(10)
```

```python
stack_df=stack_df[["year","lok_sabha_number","gender","total_electors","electors_participated","units","note"]]
stack_df.head(10)
```

```python
stack_df = stack_df.applymap(lambda x : x.strip() if isinstance(x,str) else x)
stack_df.head(5)
```

```python
stack_df.sort_values(by="year", ascending=False, inplace=True)
stack_df.head(10)
```

```python
stack_df.to_csv("/content/drive/MyDrive/Project1_processed/Table 5.2/output.csv", index=False)
```
