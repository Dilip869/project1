```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('/content/drive/MyDrive/Project1_processed/Chapter5.xlsx','5.1')
df.head(5)
```

```python
df.drop(columns=df.columns[7:], inplace=True)
df.drop(index=[df.index[0],df.index[1],df.index[3],df.index[5]], inplace=True)
df.head(10)
```

```python
df.iloc[0]=df.iloc[0,:].ffill()
df.head(5)
```

```python
#creating a new Multiindex columns
new_columns=pd.MultiIndex.from_arrays([df.iloc[0].values,df.iloc[1].values])
new_columns
```

```python
df.drop(index=df.index[:2], inplace=True)
df.head(5)
```

```python
df.columns=new_columns
df.head(10)
```

```python
df.set_index(df.columns[0],drop=True, inplace=True)
df.head(5)
```

```python
stack_df=df.stack(level=1).reset_index()
stack_df.head(10)
```

```python
stack_df.rename(columns={stack_df.columns[0]:"year",stack_df.columns[1]:"minister_type",stack_df.columns[2]:"number_of_ministers",stack_df.columns[3]:"number_of_women_ministers"}, inplace=True)
stack_df.head(5)
```

```python
stack_df=stack_df.assign(units="number_of_ministers in absolute number, number_of_women_ministers in absolute number")
stack_df=stack_df.assign(state="All India")
stack_df.head(10)
```

```python
stack_df=stack_df[["year","state","minister_type","number_of_ministers","number_of_women_ministers","units"]]
stack_df.head(5)
```

```python
stack_df.sort_values(by=["year"], ascending=False, inplace=True)
stack_df.head(10)
```

```python
stack_df.to_csv('/content/drive/MyDrive/Project1_processed/Table 5.1/output.csv', index=False)
```
