```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel("/content/drive/MyDrive/Project1_processed/Chapter5.xlsx","5.4")
df.head(5)
```

```python
df.drop(index=[df.index[0],df.index[1], df.index[3],df.index[5]], inplace=True)
df.head(5)
```

```python
df.loc[df.index[1],df.columns[1]]="year"
df.loc[df.index[1],df.columns[2]]="lok_sabha_tenure"
df.loc[df.index[1],df.columns[3]]="seats_available"
df.loc[df.index[1],df.columns[4]]="total_contested"
df.loc[df.index[1],df.columns[6]]="Total Contested"
df.loc[df.index[1],df.columns[7]]="Total Elected"
df.loc[df.index[1],df.columns[9]]="Total Contested"
df.loc[df.index[1],df.columns[10]]="Total Elected"
df.loc[df.index[1],df.columns[12]]="lok_sabha_number"
df.head(5)
```

```python
df.drop(columns=[df.columns[0],df.columns[5],df.columns[8],df.columns[11]], inplace=True)
df.head(5)
```

```python
df = df.applymap(lambda x : x.strip() if isinstance(x,str) else x)
```

```python
df.set_index(df.columns[8], inplace=True)
df.head(5)
```

```python
df.iloc[0]=df.iloc[0].ffill().values
df.iloc[2]=df.iloc[2].ffill().values
df.head(5)
```

```python
new_columns=pd.MultiIndex.from_arrays([df.iloc[0].values,df.iloc[1].values])
```

```python
df.drop(index=[df.index[0],df.index[1]], inplace=True)
df.head(5)
```

```python
df.columns=new_columns
df.head(5)
```

```python
df.reset_index(drop=False, inplace=True)
df.head(5)
```

```python
df.set_index([df.columns[0],df.columns[1],df.columns[2],df.columns[3],df.columns[4]],drop=True, inplace=True)
df.head(5)
```

```python
stack_df=df.stack(level=0).reset_index()
stack_df.head(5)
```

```python
stack_df.rename(columns={stack_df.columns[0]:"lok_sabha_number",stack_df.columns[1]:"year",
                         stack_df.columns[2]:"lok_sabha_tenure",stack_df.columns[3]:"total_seats_available",
                         stack_df.columns[4]:"total_contested", stack_df.columns[5]:"gender"}, inplace=True)
stack_df.head(10)
```

```python
melt_df=stack_df.melt(id_vars=[stack_df.columns[0],stack_df.columns[1],stack_df.columns[2],stack_df.columns[3],
                               stack_df.columns[4],stack_df.columns[5]], var_name="category", value_name="value")
melt_df.head(10)
```

```python
melt_df=melt_df.assign(units="total_seat_available in absolute numbers, total_contested in absolute numbers, value in absolute numbers")
```

```python
melt_df["note1"]=melt_df["lok_sabha_number"].apply(lambda x:"lok_sabha_number:Sex-wise break up of candidates are not available" if "First" in x else "")
melt_df["note2"]=melt_df["category"].apply(lambda x: "category:Total number of contestants include the number of candidates elected unopposed, if any" if "contesting" in x else "")
melt_df["value"]=melt_df["value"].apply(lambda x: "" if "Sex-wise break up of candidates are not available" in str(x) else x)
melt_df.head(10)
```

```python
melt_df['note']=melt_df['note1']+melt_df['note2']
```

```python
melt_df.drop(['note1','note2'], axis=1, inplace=True)
melt_df.head(10)
```

```python
melt_df=melt_df[['year','lok_sabha_number','lok_sabha_tenure','total_seats_available','total_contested',
                 'gender','category','value','units','note']]
melt_df.head(10)
```

```python
melt_df=melt_df.sort_values(by=["year"], ascending=False)
melt_df.head(10)
```

```python
melt_df.to_csv("/content/drive/MyDrive/Project1_processed/Table 5.4/output.csv", index=False)
```
