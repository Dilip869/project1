```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('/content/drive/MyDrive/Project1_processed/Chapter5.xlsx','5.11')
df.head(5)
```

```python
df.dropna(thresh=abs(df.shape[1]*0.5), axis=0, inplace=True)
df.dropna(how="all", axis=1, inplace=True)
df.head(5)
```

```python
df.drop(columns= df.columns[0], inplace=True)
df.head(5)
```

```python
df.drop(index=[df.index[0],df.index[2]], inplace=True)
```

```python
df.columns=df.iloc[0,:].values
df.head(5)
```

```python
df.drop(index=df.index[0], inplace=True)
df.head(5)
```

```python
melt_df=df.melt(id_vars=df.columns[3], var_name="region",value_name="value")
melt_df.head(5)
```

```python
melt_df.rename(columns={melt_df.columns[0]:"state"}, inplace=True)
```

```python
melt_df=melt_df.assign(units="value in percentage")
melt_df.head(10)
```

```python
note_dict={"-": "corresponding to a category implies that there are no sample persons in that particular category."}
```

```python
for each_note in note_dict:
  melt_df["note"]=melt_df["value"].apply(lambda x: f"value:{note_dict[each_note]}" if f"{each_note}" in str(x) else "")
melt_df.head(10)
```

```python
melt_df=melt_df.assign(fiscal_year="2019-20")
melt_df.head(5)
```

```python
melt_df["value"]=melt_df["value"].apply(lambda x:"" if "-" in str(x) else x)
```

```python
melt_df=melt_df[["fiscal_year","state","region","value","units","note"]]
```

```python
new_df=pd.read_excel('/content/drive/MyDrive/Project1_processed/data_dictionaries.xlsx')
new_df.head(5)
```

```python
ref_df=pd.DataFrame(melt_df["state"].unique())
ref_df.head(5)
```

```python
list1=new_df["state"].to_list()
list2=ref_df[ref_df.columns[0]].to_list()
```
